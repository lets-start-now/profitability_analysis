```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

all_sheets = pd.read_excel('Тестовое_задание_для_Junior_аналитика.xlsx', sheet_name = None)
df1 = all_sheets['Зарплата_и_премии']
df2 = all_sheets['Выручка_по_проектам']
df3 = all_sheets['Распределение_времени']
df = pd.DataFrame(df1['Пользователь'])
df['obshiy rashod na kvartal'] = 3 * df1['Базовая зарплата (мес)']+df1['Премия (разовая)']
df.head()
```

```python
plt.bar(df['Пользователь'], df['obshiy rashod na kvartal']/1000000, color = 'green')
plt.title('Obshaya summa zatrat na kajdogo sotrudnika za kvartal')
plt.xlabel('Sotrudniki')
plt.ylabel('Zatrata (M)')
plt.xticks(rotation = 45)
plt.show()
```

```python
alpha_time = df3['Alpha'].sum() 
beta_time = df3['Beta'].sum() 
gamma_time = df3['Gamma'].sum() 
df['viruchka po Alpha (%)'] = (df3['Alpha']/alpha_time * 100)
df['viruchka po Beta (%)'] = round(df3['Beta']/beta_time *100)
df['viruchka po Gamma (%)'] = round(df3['Gamma']/gamma_time *100)
df
```

```python
plt.figure(figsize=(10, 6))
bar1 = plt.bar(df['Пользователь'], df['viruchka po Alpha (%)'], label='Alpha', color='#7FB3D5')
bar2 = plt.bar(df['Пользователь'], df['viruchka po Beta (%)'], bottom=df['viruchka po Alpha (%)'], label='Beta',  color='#82E0AA')
bar3 = plt.bar(df['Пользователь'], df['viruchka po Gamma (%)'],
               bottom=df['viruchka po Alpha (%)'] + df['viruchka po Beta (%)'], label='Gamma', color='#85C1E9')

plt.ylabel('Virucka (%)')
plt.title('Raspredeleniye virucki (%)')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
df['Rentabelnost'] = ((df['viruchka po Alpha (%)']*df2['Зачисление от клиента'][0]+df['viruchka po Beta (%)']*df2['Зачисление от клиента'][1]+df['viruchka po Gamma (%)']*df2['Зачисление от клиента'][2])/(100 * df['obshiy rashod na kvartal'])).astype(float)
df
```

```python
plt.pie(df['Rentabelnost'],labels = df['Пользователь'], autopct='%1.1f%%')
plt.show()
```
